```python
inc_name=input('請輸入公司名稱')
while True:
  year=input('請輸入年份')
  if year.isdecimal():
    break
  else:
    print('請以數字形式輸入年份')
    continue

total_assets=0
tlasq=1
while total_assets!=tlasq:
  n=1
  while n!=0:
    n=0
    list_current_assets=input('請依序輸入CASH,A/R,INVENTORY,PREPAID EXPENSE,INVESTMENT，並如前述格式以逗號","分隔').split(',')
    if len(list_current_assets)==5:
      for a in list_current_assets:
        if a.isdecimal() is False:
          print(f'數值{a}輸入有誤')
          n+=1
    else:
      print('請輸入五筆資料，並以逗號分隔')
      n+=1
      print('請重新輸入資料')

  current_assets=0
  for a in list_current_assets:
    current_assets+=int(a)


  n=1
  while n!=0:
    n=0
    list_property_and_equipment=input('請依序輸入LAND,BUILDINGS AND IMPROVEMENT,EQUIPMENT,LESS ACCUMULATED DEPRECIATION)').split(',')
    if len(list_property_and_equipment)==4:
      for a in list_property_and_equipment:
        if a.isdecimal() is False:
          print(f'數值{a}輸入有誤')
          n+=1
    else:
      print('請輸入四筆資料，並以逗號分隔')
      n+=1
      print('請重新輸入資料')

  property_and_equipment=0
  for a in list_property_and_equipment:
    property_and_equipment+=int(a)
  property_and_equipment-=int(list_property_and_equipment[3])*2


  n=1
  while n!=0:
    n=0
    list_other_assets=input('請依序輸入INTANGIBLE ASSETS, LESS ACCUMULATED AMORTIZATION').split(',')
    if len(list_other_assets)==2:
      for a in list_other_assets:
        if a.isdecimal() is False:
          print(f'數值{a}輸入有誤')
          n+=1
    else:
      print('請輸入兩筆資料，並以逗號分隔')
      n+=1
      print('請重新輸入資料')

  other_assets=int(list_other_assets[0])-int(list_other_assets[1])


  total_assets=str(current_assets+property_and_equipment+other_assets)


  n=1
  while n!=0:
    n=0
    list_current_liabilities=input('請依序輸入A/P,N/P,ACCRUED INTEREST,DEFERRED REVENUE').split(',')
    if len(list_current_liabilities)==4:
      for a in list_current_liabilities:
        if a.isdecimal() is False:
          print(f'數值{a}輸入有誤')
          n+=1
    else:
      print('請輸入四筆資料，並以逗號分隔')
      n+=1
      print('請重新輸入資料')

  current_liabilities=0
  for a in list_current_liabilities:
    current_liabilities+=int(a)

  while True:
    long_term_debt=input('請輸入long term debt')
    if long_term_debt.isdecimal():
      break
    else:
      print('輸入有誤')
      continue
  total_liabilities=str(current_liabilities+int(long_term_debt))
  print(total_liabilities)

  n=1
  while n!=0:
    n=0
    list_shareholders_equity=input('請依序輸入COMMON STOCK, ADDITIONAL PAID-IN CAPITAL, RETAINED EARNINGS, TREASURY STOCKS').split(',')
    if len(list_shareholders_equity)==4:
      for a in list_shareholders_equity:
        if a.isdecimal() is False:
          print(f'數值{a}輸入有誤')
          n+=1
    else:
      print('請輸入四筆資料，並以逗號分隔')
      n+=1
      print('請重新輸入資料')
  print(list_shareholders_equity)
  shareholders_equity=0
  for a in list_shareholders_equity:
    shareholders_equity+=int(a)
  shareholders_equity-=int(list_shareholders_equity[3])*2
  print(shareholders_equity)

  tlasq=str(int(total_liabilities)+shareholders_equity)
  print(tlasq)

  if total_assets!=tlasq:
    print('輸入值有誤,total assets不等於total liabilities and shareholders equity')


print(f'{inc_name:^114s}')
print(f'{"BALANCE SHEET":^114s}')
print(f'{("December 31,"+year):^114s}')
print(f'{"assets":<60s}{"liabilities and shareholders equity":<54s}')
print(f'{"current assets":<60s}{"current liabilities":<54s}')
print(f'{" "*6}{"cash and cash equivalents":<39s}${list_current_assets[0]:>8s}{" "*12}{"accounts payable":<39s}${list_current_liabilities[0]:>8s}')
print(f'{" "*6}{"accounts receivable":<40s}{list_current_assets[1]:>8s}{" "*12}{"notes payable":<40s}{list_current_liabilities[1]:>8s}')
print(f'{" "*6}{"inventory":<40s}{list_current_assets[2]:>8s}{" "*12}{"accrued expenses":<40s}{list_current_liabilities[2]:>8s}')
print(f'{" "*6}{"prepaid expense":<40s}{list_current_assets[3]:>8s}{" "*12}{"deferred revenue":<40s}{list_current_liabilities[3]:>8s}')
print(f'{" "*6}{"investments":<40s}{list_current_assets[4]:>8s}{" "*6}{"-"*54}')
print(f'{"-"*54}{" "*12}{"total current liabilities":<40s}{current_liabilities:>8d}')
print(f'{" "*6}{"total current assets":<40s}{current_assets:>8d}{" "*6}{"-"*54}')
print(f'{"-"*54}{" "*60}')
print(f'{" "*66}{"long-term debt":<40s}{long_term_debt:>8s}')
print(f'{"property and equipment":<114s}')
print(f'{" "*6}{"land":<40s}{list_property_and_equipment[0]:>8s}{" "*52}{"-"*8}')
print(f'{" "*6}{"buildings and improvements":<40s}{list_property_and_equipment[1]:>8s}{" "*6}{"total liabilities":<46s}{total_liabilities:>8s}')
print(f'{" "*6}{"equipment":<40s}{list_property_and_equipment[2]:>8s}{" "*52}{"-"*8}')
print(f'{" "*6}{"less auumulated depreciation":<40s}{"("+list_property_and_equipment[3]+")":>8s}{" "*60}')
print(f'{" "*60}{"shareholders equity":<54s}')
print(f'{"other assets":<66s}{"common stock":<40s}{list_shareholders_equity[0]:>8s}')
print(f'{" "*6}{"intangible assets":<40s}{list_other_assets[0]:>8s}{" "*12}{"additional paid-in capital":<40}{list_shareholders_equity[1]:>8s}')
print(f'{" "*6}{"less accumulated amortization":<40s}{"("+list_other_assets[1]+")":>8s}{" "*12}{"retained earnings":<40s}{list_shareholders_equity[2]:>8s}')
print(f'{" "*66}{"treasury stock":<40s}{"("+list_shareholders_equity[3]+")":>8s}')
print(f'{" "*46}{"-"*8}{" "*60}')
print(f'{"total assets":<45s}${total_assets:>8s}{" "*52}{"-"*8}')
print(f'{" "*46}{"="*8}{" "*6}{"total liabilities and shareholders equity":<45s}${tlasq:>8s}')
print(f'{" "*106}{"="*8}')
```
