```python
import numpy as np
from numpy import random as nprd

def sampling(N):
    d=nprd.rand(N)<0.5
    z=nprd.randn(N)
    x=np.array([z[i]+5 if d[i] else z[i]-5 for i in range(N)])
    return x

N=[2,5,10,50,100,1000]
M=2000
MEANS=[]
for n in N:
    mean_x=np.zeros(M)
    for k in range(M):
        x=sampling(n)
        mean_x[k]=np.mean(x)/np.sqrt(26/n)
    MEANS.append(mean_x)

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 8.0)

x=sampling(1000)
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Histogram of Mixture Normal Distribution')
plt.hist(x,bins=40)
plt.show()

ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,3,4)
ax5 = plt.subplot(2,3,5)
ax6 = plt.subplot(2,3,6)

x=np.linspace(-5,5,25)
d=[1.0/np.sqrt(2*np.pi)*np.exp(-i**2/2) for i in x]

def plot_hist_density(ax,data,N):
    ax.hist(data,bins=25,density=True)
    ax.plot(x,d)
    ax.set_title(r'Histogram of $\bar{x}$ : N=%d' % N)

plot_hist_density(ax1,MEANS[0],N[0])
plot_hist_density(ax2,MEANS[1],N[1])
plot_hist_density(ax3,MEANS[2],N[2])
plot_hist_density(ax4,MEANS[3],N[3])
plot_hist_density(ax5,MEANS[4],N[4])
plot_hist_density(ax6,MEANS[5],N[5])

plt.show()
```
